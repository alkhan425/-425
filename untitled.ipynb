```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv("https://drive.google.com/uc?id=1D-Ypp-ooR-fjmySz0NU1FgWJ9u6Apt2W")

# Create a boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental_education_level', y='exam_score', data=df)
plt.xticks(rotation=45)
plt.title('Distribution of Exam Scores by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.show()

# Calculate average scores by parental education level
avg_scores = df.groupby('parental_education_level')['exam_score'].agg(['mean', 'count', 'std']).round(2)
avg_scores = avg_scores.sort_values('mean', ascending=False)

print("\
Statistical Summary by Parental Education Level:")
print(avg_scores)

# Perform one-way ANOVA to test if differences are statistically significant
from scipy import stats

education_groups = [group for _, group in df.groupby('parental_education_level')['exam_score']]
f_statistic, p_value = stats.f_oneway(*education_groups)

print("\
One-way ANOVA Test Results:")
print(f"F-statistic: {f_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset

df = pd.read_csv("https://drive.google.com/uc?id=1D-Ypp-ooR-fjmySz0NU1FgWJ9u6Apt2W")

# Check unique values if extracurricular participation is categorical/numerical
print('Unique values for extracurricular participation:', df['extracurricular_participation'].unique())

# Assuming extracurricular participation is a categorical variable (e.g., Yes/No or levels), we visualize with a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='extracurricular_participation', y='exam_score', data=df)
plt.title('Exam Scores by Extracurricular Participation')
plt.xlabel('Extracurricular Participation')
plt.ylabel('Exam Score')
plt.show()

# Alternatively, if extracurricular participation is numeric (e.g., frequency of participation), use scatterplot
if df['extracurricular_participation'].dtype in ['int64', 'float64']:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='extracurricular_participation', y='exam_score')
    plt.title('Exam Scores vs Extracurricular Participation')
    plt.xlabel('Extracurricular Participation (Numeric Value)')
    plt.ylabel('Exam Score')
    plt.show()

# Compute statistical summary
extra_summary = df.groupby('extracurricular_participation')['exam_score'].agg(['mean', 'count', 'std']).round(2)
print('\
Statistical Summary for Exam Scores by Extracurricular Participation:')
print(extra_summary)

# If extracurricular participation is binary (e.g., Yes/No), perform t-test
from scipy import stats

values = df['extracurricular_participation'].unique()
if len(values) == 2:
    group1 = df[df['extracurricular_participation'] == values[0]]['exam_score']
    group2 = df[df['extracurricular_participation'] == values[1]]['exam_score']
    t_stat, p_val = stats.ttest_ind(group1, group2)
    print('\
T-test for difference in exam scores between groups:')
    print('T-statistic:', t_stat)
    print('p-value:', p_val)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Load the dataframe (assuming it's already loaded, but just in case)
df = pd.read_csv("https://drive.google.com/uc?id=1D-Ypp-ooR-fjmySz0NU1FgWJ9u6Apt2W")

# Let's create a correlation matrix for study related variables: 'study_hours_per_day', 'social_media_hours', 'netflix_hours', and 'exam_score'
study_vars = ['study_hours_per_day', 'social_media_hours', 'netflix_hours', 'exam_score']
corr_matrix = df[study_vars].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation between Study-Related Variables and Exam Scores')
plt.tight_layout()
plt.show()

# Scatter plot and regression for study_hours_per_day vs exam_score
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='study_hours_per_day', y='exam_score', scatter_kws={'alpha':0.6})
plt.title('Relationship between Daily Study Hours and Exam Score')
plt.xlabel('Daily Study Hours')
plt.ylabel('Exam Score')
plt.show()

# Group analysis: average exam_score by bins of study_hours_per_day
study_bins = pd.cut(df['study_hours_per_day'], bins=[0,2,4,6,8,10,15], labels=['0-2','2-4','4-6','6-8','8-10','10+'])
avg_grades = df.groupby(study_bins)['exam_score'].agg(['mean','count','std']).round(2)
print("\
Average Exam Scores by Study Hours Per Day:")
print(avg_grades)

# Calculate correlation coefficient and p-value for study_hours_per_day vs exam_score
corr_coef, p_val = stats.pearsonr(df['study_hours_per_day'], df['exam_score'])
print("\
Correlation between daily study hours and exam score:")
print(f"Correlation coefficient: {corr_coef:.4f}")
print(f"P-value: {p_val:.4e}")

# Additionally, let's explore social_media_hours and netflix_hours effects
# Scatter plot for social_media_hours vs exam_score
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='social_media_hours', y='exam_score', scatter_kws={'alpha':0.6}, color='green')
plt.title('Relationship between Social Media Hours and Exam Score')
plt.xlabel('Social Media Hours')
plt.ylabel('Exam Score')
plt.show()

# Scatter plot for netflix_hours vs exam_score
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='netflix_hours', y='exam_score', scatter_kws={'alpha':0.6}, color='purple')
plt.title('Relationship between Netflix Hours and Exam Score')
plt.xlabel('Netflix Hours')
plt.ylabel('Exam Score')
plt.show()

# Calculate correlations for these variables
corr_social, p_social = stats.pearsonr(df['social_media_hours'], df['exam_score'])
corr_netflix, p_netflix = stats.pearsonr(df['netflix_hours'], df['exam_score'])

print("\
Correlation between social media hours and exam score:")
print(f"Coefficient: {corr_social:.4f}, P-value: {p_social:.4e}")

print("\
Correlation between Netflix hours and exam score:")
print(f"Coefficient: {corr_netflix:.4f}, P-value: {p_netflix:.4e}")
```
